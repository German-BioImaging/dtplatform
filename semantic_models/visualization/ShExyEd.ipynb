```python
schema = json.loads(shex)
```

```python
prefixes = {}
prefixes["http://subject.deltatissue.org/subject"] = "subject:"
prefixes["http://www.w3.org/2001/XMLSchema#"] = "xsd:"
prefixes["http://purl.obolibrary.org/obo/"] = "obo:"
prefixes["http://purl.org/dc/terms/"] = "dct:"
prefixes["http://www.w3.org/1999/02/22-rdf-syntax-ns#"] = "rdf:"
prefixes["http://purl.org/dc/elements/1.1/"] = "dce:"
```

```python
import uuid
level = 0
nodes = []
edges = []

for shape in schema["shapes"]:
    print(str(level), shape["id"])
    node = dict()
    node["id"] = shape["id"]
    label = shape["id"] 
    for prefix in prefixes.keys():
       label = label.replace(prefix, prefixes[prefix])
    node["properties"] = {"type": 0, "label": label}
    nodes.append(node)
    
    for expression in shape["expression"]["expressions"]:
        print(type(expression["valueExpr"]))
        if type(expression["valueExpr"]) is dict:
            if "datatype" in expression["valueExpr"].keys():
                print(shape["id"], expression["predicate"], expression["valueExpr"]["datatype"])
                node = dict()
                node_uuid = uuid.uuid4().hex
                node["id"] = node_uuid
                label = expression["valueExpr"]["datatype"] 
                for prefix in prefixes.keys():
                   label = label.replace(prefix, prefixes[prefix])
                node["properties"] = {"type": 1, "label": label}
                nodes.append(node)
                
                edge = dict()
                edge["id"] = expression["predicate"]
                edge["start"] = shape["id"]
                edge["end"] = node_uuid
                edge["directed"] = True
                label = expression["predicate"]
                for prefix in prefixes.keys():
                   label = label.replace(prefix, prefixes[prefix])
                edge["properties"] = {"label": label}
                edges.append(edge)
                
            if "values" in expression["valueExpr"].keys():
                for value in expression["valueExpr"]["values"]:
                    if type(value) is dict:
                        value = value["value"]
                    print(shape["id"], expression["predicate"], value)
                    node = dict()
                    node["id"] = value
                    node["properties"] = {"type": 2, "label": value}
                    nodes.append(node)
                    
                    edge = dict()
                    edge["id"] = expression["predicate"]
                    edge["start"] = shape["id"]
                    edge["end"] = value
                    label = expression["predicate"]
                    for prefix in prefixes.keys():
                       label = label.replace(prefix, prefixes[prefix])
                    edge["properties"] = {"label": label}
                    edges.append(edge)        
                    
        if type(expression["valueExpr"]) is str:
            print(shape["id"], expression["predicate"], expression["valueExpr"])
            node = dict()
            node["id"] = expression["valueExpr"]
            node["properties"] = {"type": 0, "label": expression["valueExpr"]}
            nodes.append(node)
            
            edge = dict()
            edge["id"] = expression["predicate"]
            edge["start"] = shape["id"]
            edge["end"] = expression["valueExpr"] 
            label = expression["predicate"]
            for prefix in prefixes.keys():
                label = label.replace(prefix, prefixes[prefix])
            edge["properties"] = {"type": 0, "label": label}
            edges.append(edge)
            
```

```python
schemaViz = GraphWidget()

colors = [ "#ffc914", "#0b7189", "#ff6c00", '#76b041', "#17bebb"]
def custom_node_color_mapping(index, node):
    """assign colors based on type"""
    return colors[node['properties']['type']]

schemaViz.set_node_color_mapping(custom_node_color_mapping)

schemaViz.nodes = nodes
schemaViz.edges = edges
```
